````python
import re
import time
import uuid
import os
import psutil
def tokenize(statement):
  """Tokenizes a statement into individual words.

  Args:
    statement: The input statement as a string.

  Returns:
    A list of strings, where each string is a word from the statement.
  """
  # Remove punctuation and convert to lowercase
  statement = re.sub(r'[^\w\s]', '', statement).lower()
  # Split the statement into words
  words = statement.split()
  return words

# Test case
input_statement = "Hello, how are you?"
output_tokens = tokenize(input_statement)
print(output_tokens)
print("---------------------------------------")
# Time taken
start_time = time.time()
# MAC Address
mac = uuid.getnode()
mac_address = ':'.join(("%012X" % mac)[i:i+2] for i in range(0, 12, 2))
print(mac_address)
end_time = time.time()
time_taken = end_time - start_time
# Memory usage
process = psutil.Process(os.getpid())
memory_usage = process.memory_info().rss / (1024 * 1024)  # in MB
# Result
print(mac_address)
print(f"Time taken: {time_taken} seconds")
print(f"Memory usage: {memory_usage} MB")
print("---------------------------------------")
````

````python
import time
import uuid
import os
import psutil
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Download required NLTK data
nltk.download('punkt')
nltk.download('punkt_tab') # Download punkt_tab for sentence tokenization

def calculate_semantic_similarity(sentence1, sentence2):
    # Tokenize sentences
    tokens1 = nltk.word_tokenize(sentence1.lower())
    tokens2 = nltk.word_tokenize(sentence2.lower())

    # Create TF-IDF vectors
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([sentence1, sentence2])

    # Calculate cosine similarity
    similarity_score = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]
    return similarity_score

# Example usage
sentence1 = "This is a sample sentence."
sentence2 = "This sentence is just a sample."
similarity = calculate_semantic_similarity(sentence1, sentence2)
print(f"Semantic Similarity: {similarity:.2f}")  # Output will be around 0.8
print("---------------------------------------")
# Time taken
start_time = time.time()
# MAC Address
mac = uuid.getnode()
mac_address = ':'.join(("%012X" % mac)[i:i+2] for i in range(0, 12, 2))
print(mac_address)
end_time = time.time()
time_taken = end_time - start_time
# Memory usage
process = psutil.Process(os.getpid())
memory_usage = process.memory_info().rss / (1024 * 1024)  # in MB
# Result
print(mac_address)
print(f"Time taken: {time_taken} seconds")
print(f"Memory usage: {memory_usage} MB")
print("---------------------------------------")
````

````python
import time
import uuid
import os
import psutil
from sklearn.feature_extraction.text import TfidfVectorizer

import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = text.split()
    words = [word for word in words if word not in stop_words]
    # Join the words back into a string
    text = ' '.join(words)
    return text

# Example usage
input_text = "This is a sample text. It contains punctuation and stopwords."
processed_text = preprocess_text(input_text)
print(processed_text) # Output: sample text contains
print("---------------------------------------")
# Time taken
start_time = time.time()
# MAC Address
mac = uuid.getnode()
mac_address = ':'.join(("%012X" % mac)[i:i+2] for i in range(0, 12, 2))
print(mac_address)
end_time = time.time()
time_taken = end_time - start_time
# Memory usage
process = psutil.Process(os.getpid())
memory_usage = process.memory_info().rss / (1024 * 1024)  # in MB
# Result
print(mac_address)
print(f"Time taken: {time_taken} seconds")
print(f"Memory usage: {memory_usage} MB")
print("---------------------------------------")
````

````python
import time
import uuid
import os
import psutil

import numpy as np

class SkipGram:
    def __init__(self, vocab_size, embedding_dim, context_size):
        self.vocab_size = vocab_size
        self.embedding_dim = embedding_dim
        self.context_size = context_size

        # Initialize embedding matrix with random values
        self.W1 = np.random.randn(vocab_size, embedding_dim)
        self.W2 = np.random.randn(embedding_dim, vocab_size)

    def one_hot_encode(self, word_index):
        vector = np.zeros(self.vocab_size)
        vector[word_index] = 1
        return vector

    def forward(self, target_word_index):
        # One-hot encode the target word
        x = self.one_hot_encode(target_word_index)

        # Hidden layer (embedding lookup)
        h = np.dot(x, self.W1)

        # Output layer (softmax)
        u = np.dot(h, self.W2)
        y_pred = self.softmax(u)

        return y_pred, h, x

    def softmax(self, x):
        e_x = np.exp(x - np.max(x)) # subtract max for numerical stability
        return e_x / e_x.sum(axis=0)

    def backpropagation(self, error, h, x, learning_rate):
        dW2 = np.outer(h, error)
        dW1 = np.outer(x, np.dot(self.W2, error))

        # Update weights
        self.W1 -= learning_rate * dW1
        self.W2 -= learning_rate * dW2


    def train(self, data, epochs, learning_rate):
        for epoch in range(epochs):
          loss = 0
          for target_word_index, context_word_indices in data:
              # Forward Pass
              y_pred, h, x = self.forward(target_word_index)

              # Calculate error for each context word (simplified error calculation, replace with negative sampling or hierarchical softmax for better performance)
              for context_word_index in context_word_indices:
                  error = y_pred.copy()
                  error[context_word_index] -= 1  # Error calculation (simplified)
                  self.backpropagation(error, h, x, learning_rate)
                  loss += -np.log(y_pred[context_word_index]) #Simple Cross Entropy

          print(f"Epoch: {epoch+1}, Loss: {loss/len(data)}")  # Average Loss per sentence



# Example Usage (replace with your own data)
vocab_size = 10000
embedding_dim = 100
context_size = 2
learning_rate = 0.01
epochs = 100

# Sample Data (list of (target_word_index, [context_word_indices]) pairs)
data = [(1, [2, 3]), (2, [1, 4]),(3,[1,5])]


model = SkipGram(vocab_size, embedding_dim, context_size)
model.train(data, epochs, learning_rate)

# Time taken, MAC Address, Memory usage
start_time = time.time()
mac = uuid.getnode()
mac_address = ':'.join(("%012X" % mac)[i:i+2] for i in range(0, 12, 2))
end_time = time.time()
time_taken = end_time - start_time
process = psutil.Process(os.getpid())
memory_usage = process.memory_info().rss / (1024 * 1024)
print("---------------------------------------")
print(mac_address)
print(f"Time taken: {time_taken} seconds")
print(f"Memory usage: {memory_usage} MB")
print("---------------------------------------")
````

````python
import time
import uuid
import os
import psutil
import numpy as np

class CBOW:
    def __init__(self, vocab_size, embedding_dim, context_size):
        self.vocab_size = vocab_size
        self.embedding_dim = embedding_dim
        self.context_size = context_size

        # Initialize embedding matrices
        self.W1 = np.random.randn(vocab_size, embedding_dim)
        self.W2 = np.random.randn(embedding_dim, vocab_size)

    def one_hot_encode(self, word_index):
        vector = np.zeros(self.vocab_size)
        vector[word_index] = 1
        return vector

    def forward(self, context_word_indices):
        # Average the embeddings of context words
        context_embeddings = np.mean([self.W1[word_index] for word_index in context_word_indices], axis=0)

        # Output layer (softmax)
        u = np.dot(context_embeddings, self.W2)
        y_pred = self.softmax(u)

        return y_pred, context_embeddings

    def softmax(self, x):
        e_x = np.exp(x - np.max(x))  # subtract max for numerical stability
        return e_x / e_x.sum(axis=0)

    def backpropagation(self, error, context_embeddings, context_word_indices, learning_rate):
      # Update W2
      dW2 = np.outer(context_embeddings, error)
      self.W2 -= learning_rate * dW2

      # Update W1 (context word embeddings)
      for word_index in context_word_indices:
          dW1 = np.outer(self.one_hot_encode(word_index), np.dot(self.W2, error) / len(context_word_indices))
          self.W1 -= learning_rate * dW1

    def train(self, data, epochs, learning_rate):
        for epoch in range(epochs):
            loss = 0
            for context_word_indices, target_word_index in data:
                # Forward pass
                y_pred, context_embeddings = self.forward(context_word_indices)

                # Calculate error (simplified error calculation)
                error = y_pred.copy()
                error[target_word_index] -= 1

                # Backpropagation
                self.backpropagation(error, context_embeddings, context_word_indices, learning_rate)

                loss += -np.log(y_pred[target_word_index])  # Cross-entropy loss

            print(f"Epoch: {epoch + 1}, Loss: {loss / len(data)}")

# Example usage (replace with your own data):
vocab_size = 10000
embedding_dim = 100
context_size = 2
learning_rate = 0.01
epochs = 100

# Sample data (list of ([context_word_indices], target_word_index) pairs)
data = [([2, 3], 1), ([1, 4], 2),([1,5],3)]

model = CBOW(vocab_size, embedding_dim, context_size)
model.train(data, epochs, learning_rate)
# Time taken, MAC Address, Memory usage
start_time = time.time()
mac = uuid.getnode()
mac_address = ':'.join(("%012X" % mac)[i:i+2] for i in range(0, 12, 2))
end_time = time.time()
time_taken = end_time - start_time
process = psutil.Process(os.getpid())
memory_usage = process.memory_info().rss / (1024 * 1024)
print("---------------------------------------")
print(mac_address)
print(f"Time taken: {time_taken} seconds")
print(f"Memory usage: {memory_usage} MB")
print("---------------------------------------")
````

````python
import time
import uuid
import os
import psutil

import numpy as np
from collections import defaultdict

class Glove:
    def __init__(self, vocab_size, embedding_dim, X, learning_rate=0.05, epochs=100):
        self.vocab_size = vocab_size
        self.embedding_dim = embedding_dim
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.X = X  # Co-occurrence matrix
        self.W = np.random.normal(0, 0.1, (vocab_size, embedding_dim)) # Weights
        self.b = np.random.normal(0, 0.1, vocab_size) # Biases
        self.W_context = np.random.normal(0, 0.1, (vocab_size, embedding_dim)) # Weights
        self.b_context = np.random.normal(0, 0.1, vocab_size) # Biases
        self.alpha = 0.75  # Weighting exponent

    def train(self):
        for epoch in range(self.epochs):
            total_loss = 0
            for i in range(self.vocab_size):
                for j in range(self.vocab_size):
                    if self.X[i, j] > 0:
                        weight = (self.X[i, j] / 100) ** self.alpha if self.X[i,j] < 100 else 1 # Weighting function
                        diff = np.dot(self.W[i], self.W_context[j]) + self.b[i] + self.b_context[j] - np.log(self.X[i,j])
                        loss = weight * (diff ** 2)
                        total_loss += loss

                        # Update weights and biases
                        grad_w = 2 * weight * diff * self.W_context[j]
                        grad_b = 2 * weight * diff
                        grad_w_context = 2 * weight * diff * self.W[i]
                        grad_b_context = 2 * weight * diff

                        self.W[i] -= self.learning_rate * grad_w
                        self.b[i] -= self.learning_rate * grad_b
                        self.W_context[j] -= self.learning_rate * grad_w_context
                        self.b_context[j] -= self.learning_rate * grad_b_context

            print(f"Epoch: {epoch+1}, Loss: {total_loss}")



# Example usage (replace with actual data and parameters)
vocab_size = 100
embedding_dim = 50

# Create a sample co-occurrence matrix (replace with your actual matrix)
X = np.random.randint(0, 10, size=(vocab_size, vocab_size))
np.fill_diagonal(X, 0) # Ensuring words dont co-occur with themselves

model = Glove(vocab_size, embedding_dim, X)
model.train()

# Get word embeddings (combined weights for better representation)
word_embeddings = model.W + model.W_context

# Time taken, MAC Address, Memory usage
start_time = time.time()
mac = uuid.getnode()
mac_address = ':'.join(("%012X" % mac)[i:i+2] for i in range(0, 12, 2))
end_time = time.time()
time_taken = end_time - start_time
process = psutil.Process(os.getpid())
memory_usage = process.memory_info().rss / (1024 * 1024)
print("---------------------------------------")
print(mac_address)
print(f"Time taken: {time_taken} seconds")
print(f"Memory usage: {memory_usage} MB")
print("---------------------------------------")
````

````python


!pip install nltk scikit-learn pandas
import time
import uuid
import os
import psutil
import nltk
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re

# Download necessary NLTK resources (only needed once)
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text).lower()  # Remove punctuation and lowercase
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words and w.isalnum()] # Remove stopwords and non-alphanumeric
    return " ".join(tokens)

# Load job descriptions (replace with actual file path)
try:
    jobs_df = pd.read_csv('jobs_data.csv') # Replace with your file name
except FileNotFoundError:
    print("Error: 'jobs_data.csv' not found. Please upload the dataset.")
    jobs_df = pd.DataFrame() # Initialize as empty dataframe if the file is missing


def recommend_jobs(resume_text, top_n=5):
    if jobs_df.empty:
        print("No job data available for recommendations.")
        return []

    resume_text = preprocess_text(resume_text)

    # Preprocess job descriptions
    jobs_df['processed_description'] = jobs_df['description'].astype(str).apply(preprocess_text)

    # Create TF-IDF vectors
    vectorizer = TfidfVectorizer()
    all_descriptions = list(jobs_df['processed_description']) + [resume_text] # Combine for fitting
    tfidf_matrix = vectorizer.fit_transform(all_descriptions)

    # Calculate cosine similarity
    resume_tfidf = tfidf_matrix[-1]
    cosine_similarities = cosine_similarity(resume_tfidf, tfidf_matrix[:-1])

    # Get top recommendations
    related_docs_indices = cosine_similarities.argsort()[0][-top_n:][::-1]
    recommendations = jobs_df.iloc[related_docs_indices]
    return recommendations


# Example usage
resume = """Experienced software engineer with expertise in Python, Java, and cloud computing.
Proficient in machine learning algorithms and deep learning frameworks."""

recommendations = recommend_jobs(resume)

# Check if recommendations is empty using len()
if len(recommendations) > 0: # Changed from 'if not recommendations.empty:'
    print(recommendations[['title', 'company', 'description']])

# Time taken, MAC Address, Memory usage
start_time = time.time()
mac = uuid.getnode()
mac_address = ':'.join(("%012X" % mac)[i:i+2] for i in range(0, 12, 2))
end_time = time.time()
time_taken = end_time - start_time
process = psutil.Process(os.getpid())
memory_usage = process.memory_info().rss / (1024 * 1024)
print("---------------------------------------")
print(mac_address)
print(f"Time taken: {time_taken} seconds")
print(f"Memory usage: {memory_usage} MB")
print("---------------------------------------")
````
